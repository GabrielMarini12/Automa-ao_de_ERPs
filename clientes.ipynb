```python
# fakturama -> link na descrição -> ERP de exemplo que vamos usar
# pyautogui -> é a ferramenta que a gente vai usar pra automatizar ERP

import pyautogui
import pyperclip
import subprocess
import time
import pandas as pd

pyautogui.FAILSAFE = True

def encontrar_imagem(imagem):
    while not pyautogui.locateOnScreen(imagem, grayscale=True, confidence=0.9):  # reconhecimento de imagem
        time.sleep(1)
    encontrou = pyautogui.locateOnScreen(imagem, grayscale=True, confidence=0.9)
    return encontrou

def escrever_texto(texto):
    pyperclip.copy(texto)
    pyautogui.hotkey("ctrl", "v")

def direita(posicoes_imagem):
    return posicoes_imagem[0] + posicoes_imagem[2], posicoes_imagem[1] + posicoes_imagem[3]/2

# Abrir o ERP (Fakturama)
subprocess.Popen([r"C:\Program Files\Fakturama2\Fakturama.exe"])

# encontrou = None
# encontrou = x, y, largura, altura
encontrou = encontrar_imagem("fakturama.png")

# fakturama está aberto

# como fazer isso para vários produtos
tabela_produtos = pd.read_excel("Clientes.xlsx")
print(tabela_produtos)
for linha in tabela_produtos.index:
    categoria = tabela_produtos.loc[linha, "Categoria"]
    empresa = tabela_produtos.loc[linha, "Empresa"]
    nome = tabela_produtos.loc[linha, "Nome"]
    sobrenome = tabela_produtos.loc[linha, "Sobrenome"]
    rua = tabela_produtos.loc[linha, "Rua"]
    numero = tabela_produtos.loc[linha, "Numero"]
    estado = tabela_produtos.loc[linha, "Estado"]
    email = tabela_produtos.loc[linha, "Email"]
    telefone = tabela_produtos.loc[linha, "Telefone"]

    # Clicar no menu New
    encontrou = encontrar_imagem("menu_new.png")
    pyautogui.click(pyautogui.center(encontrou))

    # Clicar em New Product
    encontrou = encontrar_imagem("new_debtor.png")
    pyautogui.click(pyautogui.center(encontrou))

    encontrou = encontrar_imagem("category_cliente.png")
    pyautogui.click(direita(encontrou))
    escrever_texto(str(categoria))

    encontrou = encontrar_imagem("company_cliente.png")
    pyautogui.click(direita(encontrou))
    escrever_texto(str(empresa))

    encontrou = encontrar_imagem("name_cliente.png")
    pyautogui.click(direita(encontrou))
    escrever_texto(str(nome))

    pyautogui.press('tab')
    escrever_texto(str(sobrenome))

    encontrou = encontrar_imagem("street_cliente.png")
    pyautogui.click(direita(encontrou))
    escrever_texto(str(rua))

    encontrou = encontrar_imagem("specification_cliente.png")
    pyautogui.click(direita(encontrou))
    escrever_texto(str(numero))

    encontrou = encontrar_imagem("district_cliente.png")
    pyautogui.click(direita(encontrou))
    escrever_texto(str(estado))

    #scroll na tela
    pyautogui.scroll(-150)

    encontrou = encontrar_imagem("email_cliente.png")
    pyautogui.click(direita(encontrou))
    escrever_texto(str(email))

    encontrou = encontrar_imagem("telephone_cliente.png")
    pyautogui.click(direita(encontrou))
    escrever_texto(telefone)

    # clicar em salvar
    encontrou = encontrar_imagem("botaosave.png")
    pyautogui.click(pyautogui.center(encontrou))
```
